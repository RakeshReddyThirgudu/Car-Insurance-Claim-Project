Categorical Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Identify object-type columns
object_columns = df.select_dtypes(include=['object']).columns

# Apply label encoding to each object-type column
label_encoder = LabelEncoder()

for column in object_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df
```

Building of models

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import roc_curve, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from xgboost import XGBClassifier

X=df.drop(['is_claim'],axis=1)
y=df['is_claim']

# Normalize the data
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Logistic Regression Variations
# Standard Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Regularized Logistic Regression (L2 regularization)
lr_reg_model = LogisticRegression(penalty='l2')
lr_reg_model.fit(X_train, y_train)

# Decision Tree Variations
# Standard Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

# Gradient Boosting Decision Trees
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)

# XGBoost Classifier
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
```

Evaluation of models

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import seaborn as sns

# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 5))
    sns.heatmap(cm, annot=True, fmt=".0f", cmap="Blues", linewidths=0.5)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Function to plot ROC curve
def plot_roc_curve(y_true, y_prob, title):
    fpr, tpr, thresholds = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = {:.2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.show()

# Function to plot classification report
def plot_classification_report(y_true, y_pred, title):
    report = classification_report(y_true, y_pred, target_names=['No Claim', 'Claim'])
    print(title)
    print(report)

# Models
models = {
    'Logistic Regression': lr_model,
    'Regularized Logistic Regression': lr_reg_model,
    'Decision Tree': dt_model,
    'Random Forest': rf_model,
    'Gradient Boosting': gb_model,
    'XGBoost': xgb_model
}

# Evaluate each model
for name, model in models.items():
    # Predictions
    y_pred = model.predict(X_test)
    
    # Probabilities for ROC curve
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Plot confusion matrix
    plot_confusion_matrix(y_test, y_pred, title=f'Confusion Matrix - {name}')
    
    # Plot ROC curve
    plot_roc_curve(y_test, y_prob, title=f'ROC Curve - {name}')
    
    # Print classification report
    plot_classification_report(y_test, y_pred, title=f'Classification Report - {name}')
```
